```python
# Step 1: Acquire Data
# ----------------------

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload CSV file
print("Please upload the 'summer.csv' file...")
uploaded = files.upload()

# Read the CSV file into a DataFrame
df = pd.read_csv("summer.csv")
print("Data Loaded Successfully!")

# Display basic info
print("\nDataFrame Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Step 2: Data Transformation
# ---------------------------

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop rows with missing values (optional)
df.dropna(inplace=True)

# Normalize column names (optional step)
df.columns = df.columns.str.strip().str.lower()

# Convert year to int just in case
df['year'] = df['year'].astype(int)

# Check unique cities
unique_cities = df['city'].unique()
print(f"\nTotal unique cities that hosted Summer Olympics: {len(unique_cities)}")
print("\nList of cities:")
print(unique_cities)
```

```python
 # 1. Total Medals by Country (Gold, Silver, Bronze)
medals_by_country = df.groupby('country')['medal'].value_counts().unstack(fill_value=0)
medals_by_country['total'] = medals_by_country.sum(axis=1)

# Sorting by total medals
medals_by_country_sorted = medals_by_country.sort_values('total', ascending=False)

# Plot
plt.figure(figsize=(12,8))
medals_by_country_sorted['total'].head(20).plot(kind='bar', color='lightblue')
plt.title('Top 20 Countries by Total Medals Won')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 2. Total Medals by Year
medals_by_year = df.groupby('year')['medal'].count()

# Plot
plt.figure(figsize=(12,6))
medals_by_year.plot(kind='line', color='green')
plt.title('Total Medals Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Medals')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Top 10 Athletes by Medals Won
athletes_medals = df.groupby('athlete')['medal'].count().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12,6))
athletes_medals.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Athletes by Total Medals')
plt.xlabel('Athlete')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Medals Distribution by Gender
medals_by_gender = df.groupby('gender')['medal'].count()

# Plot
plt.figure(figsize=(8,6))
medals_by_gender.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('Medals Distribution by Gender')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# 5. Total Medals by Sport
medals_by_sport = df.groupby('sport')['medal'].count().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12,6))
medals_by_sport.head(10).plot(kind='barh', color='purple')
plt.title('Top 10 Sports by Total Medals')
plt.xlabel('Total Medals')
plt.ylabel('Sport')
plt.tight_layout()
plt.show()
```

```python
# Visualization: USA's Dominance in Swimming, Athletics, and Gymnastics
# ---------------------------------------------------------------

# Filter data for top countries (e.g., USA, Russia, Germany)
top_countries = ['USA', 'RUS', 'GER', 'CHN', 'GBR']
top_countries_data = df[df['country'].isin(top_countries)]

# Group by sport and medal type
medals_by_sport_country = top_countries_data.groupby(['country', 'sport', 'medal']).size().unstack(fill_value=0)

# Plot the stacked bar chart
medals_by_sport_country.loc['USA'].plot(kind='bar', stacked=True, figsize=(14,7), colormap='Set2')
plt.title('USA: Medal Distribution by Sport (Swimming, Athletics, Gymnastics)')
plt.xlabel('Sport')
plt.ylabel('Medals Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization: Gender Representation in Olympics Over the Years
# -------------------------------------------------------------

# Count male and female athletes per year
athletes_by_gender_year = df.groupby(['year', 'gender'])['athlete'].nunique().unstack(fill_value=0)

# Plot stacked bar chart
athletes_by_gender_year.plot(kind='bar', stacked=True, figsize=(14,7), color=['lightcoral', 'lightblue'])
plt.title('Gender Representation in Olympics Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
